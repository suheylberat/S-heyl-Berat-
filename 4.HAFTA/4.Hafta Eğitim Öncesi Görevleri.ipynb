# Problem 1
Sözlükleri kullanarak bir telefon rehberi yazın. Bu rehberde kullanıcıya kimin telefonunu görüntülemek istediğini sorun ve kullanıcının girdiği isme göre o kişinin telefon numarasını yazdırın. Proje sonunda elde edeceğiniz çıktı  şuna benzer olmalı:


Lütfen numarasını öğrenmek istediğiniz kişinin adını girin: Ahmet

Ahmet isimli kişinin numarası şu şekildedir: 0532 678 13 19

````python
# 1.Problemin Çözümünü Buraya Yazınız
telefonRehberi = {"Ahmet": "0532 678 13 19" , "Berat": "0551 219 9438" , "Michael": "0911 555 6788"}
isim = input("Lütfen numarasını öğrenmek istediğiniz kişinin adını giriniz: ")
if isim == "Ahmet" :
    print("Lütfen numarasını öğrenmek istediğiniz kişinin adını giriniz:" +isim)
    print(isim , "isimli kişinin telefon numarası:" + telefonRehberi["Ahmet"])
elif isim =="Berat" :
    print("Lütfen numarasını öğrenmek istediğiniz kişinin adını giriniz:" +isim)
    print(isim , "isimli kişinin telefon numarası: " + telefonRehberi["Berat"])
elif isim == "Michael":
    print("Lütfen numarasını öğrenmek istediğiniz kişinin adını giriniz:" +isim)
    print(isim , "isimli kişinin telefon numarası: " + telefonRehberi["Michael"])
````

# Problem 2

Daha sonra bir isim sorgulama ekranı gibi kullanıcıya kimin bilgilerini görüntülemek istediğini 
sorun ve sorgulanan kişinin ekranda gösterilmesini sağlayın. Proje sonunda elde edeceğiniz çıktı şu şekilde olmalı:


Lütfen bilgilerini görüntülemek istediğiniz çalışanın ismini girin: Mehmet Yağız

Mehmet Yağız= Memleket: Adana Yaş: 40 Görev: Direktör

````python
# 2.Problemin Çözümünü Buraya Yazınız
görüntülenen = input("Lütfen görüntülemek istediğiniz çalışanın ismini giriniz: ")
calisanlar ={"Mehmet Yağız" : "Memleket=Adana , Yaş: 40 , Görev: Direktör" ,
             "Berat Gülenc" : "Memleket = Giresun , Yaş: 18 , Görev: Öğrenci"}
if görüntülenen in calisanlar:
    print(görüntülenen , "==" , calisanlar[görüntülenen])
else:
    print(görüntülenen , "adlı kisi bulunamadı")
````

# Problem 3

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48
Not Ortalaması: 62.0


````python
# 3.Problemin Çözümünü Buraya Yazınız

ogrenciNot = {"Mehmet": [72, 66, 48] , "Berat": [95, 85, 75] , "Micheal": [60, 65, 70]}
isim = input("Lütfen notlarını görmek istediğiniz öğrencinin adını giriniz: ")
if isim == "Mehmet":
    print("Mehmet isimli öğrencinin")
    print("1.Sınav Notu: " , ogrenciNot["Mehmet"][0] )
    print("2.Sınav Notu: " , ogrenciNot["Mehmet"][1] ) 
    print("3.Sınav Notu: " , ogrenciNot["Mehmet"][2] )
    print(f"Not ortalaması: " , ((ogrenciNot["Mehmet"][0] + ogrenciNot["Mehmet"][1] + ogrenciNot["Mehmet"][2]) /3))
elif isim == "Berat":
    print("Berat isimli öğrencinin")
    print("1.Sınav Notu: " , ogrenciNot["Berat"][0] )
    print("2.Sınav Notu: " , ogrenciNot["Berat"][1] ) 
    print("3.Sınav Notu: " , ogrenciNot["Berat"][2] )
    print(f"Not ortalaması: " , ((ogrenciNot["Berat"][0] + ogrenciNot["Berat"][1] + ogrenciNot["Berat"][2]) /3))
elif isim =="Michael":
    print("Michael isimli öğrencinin")
    print("1.Sınav Notu: " , ogrenciNot["Michael"][0] )
    print("2.Sınav Notu: " , ogrenciNot["Michael"][1] ) 
    print("3.Sınav Notu: " , ogrenciNot["Michael"][2] )
    print(f"Not ortalaması: " , ((ogrenciNot["Michael"][0] + ogrenciNot["Michael"][1] + ogrenciNot["Michael"][2]) /3))
````
